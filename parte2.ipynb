# Criando o vetor TF-IDF e calculando a similaridade

Aqui será utilizado a similaridade de cosseno para avaliar o quão similares os textos são.

```python
tfidf = TfidfVectorizer(stop_words=stopwords.words('portuguese'))                                                                                                                                                                                                   
tfidf = tfidf.fit(items_title['ITE_ITEM_TITLE']) #Fit com a base de 30 mil observações
tfidf_test = tfidf.transform(items_title_test['ITE_ITEM_TITLE']) #Transform na base teste                                                                                                                                                                                                               
pairwise_similarity = tfidf_test * tfidf_test.T 
```

```python
sim_array = pairwise_similarity.toarray()
```

```python
sim_array.shape
```

```python
#items_title_test['ITE_ITEM_TITLE_2'] = items_title_test['ITE_ITEM_TITLE']
```

```python
#lista = [0]*10000
```

```python
df = pd.DataFrame(sim_array, columns=items_title_test['ITE_ITEM_TITLE'].values, index=items_title_test['ITE_ITEM_TITLE'].values)
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
df.columns[1:].tolist()
```

```python
df_final = pd.melt(df, id_vars='index', value_vars=df.columns[1:].tolist())
```

```python
df_final.rename(columns={'index':'ITE_ITEM_TITLE','variable':'ITE_ITEM_TITLE_2','value':'Similarity Score (0,1)'}, inplace=True)
```

Aqui podemos ver os textos mais similares com score decrescente, porém temos diversos textos iguais em que é claramente 1 o score de similaridade.

```python
df_final_sorted = df_final.sort_values(by='Similarity Score (0,1)', ascending=False)
df_final_sorted
```

Removendo os textos que são idênticos, temos a seguinte ordenação de similaridade.

```python
df_final['Similarity Score (0,1)'] = df_final['Similarity Score (0,1)'].round(2)
```

```python
df_final_excluindo_1 = df_final[df_final['Similarity Score (0,1)']!=1.00]
```

```python
df_final_excluindo_1_sorted = df_final_excluindo_1.sort_values(by='Similarity Score (0,1)', ascending=False)
df_final_excluindo_1_sorted
```
